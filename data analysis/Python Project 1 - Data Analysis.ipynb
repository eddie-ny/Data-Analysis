## Data Preprocessing:
#### Before analyzing the dataset, we need to preprocess it to get rid of null values and duplicates. We will use pandas library to load the csv file and perform the necessary preprocessing steps.



```python
import pandas as pd
import matplotlib as plt
import seaborn as sns
import matplotlib.pyplot as pt
%matplotlib inline
```

```python
df= pd.read_csv('data.csv')
```

```python
print(df)
```

```python
df =df.rename(columns={'volatile acidity' : 'volatile_acidity'})
df =df.rename(columns={'citric acid' : 'citric_acid'})
df =df.rename(columns={'residual sugar' : 'residual_sugar'})
df =df.rename(columns={'free sulfur dioxide' : 'free_sulfur_dioxide'})
df =df.rename(columns={'total sulfur dioxide' : 'total_sulfur_dioxide'})
df =df.rename(columns={'fixed acidity' : 'fixed_acidity'})
```

## Method for handling null values:
 
#### I will handle null values by replacing the rows that contain null values with either the mode, median or mean of each column . This approach is adopted because removing rows with null values might affect the overall analysis of the data. However, if the percentage of null values is insignificant, then alternative methods like imputation can be used.

#### Below shows the columns, the non null values in the column and the data type of each column

```python
df.info()
```

```python
df.isnull().sum()
```

#### Above is data showing the number of null values in each column

#### So the mean, mode and median of the columns with null values are found

```python
mean_FA = df.fixed_acidity.mean()
mean_CA = df.citric_acid.mean()
mean_C = df.chlorides.mean()
mean_FSD = df.free_sulfur_dioxide.mean()
mean_TSD = df.total_sulfur_dioxide.mean()
mean_D = df.density.mean()
mean_pH = df.pH.mean()
mean_Q = df.quality.mean()
```

```python
print(f'Fixed_Acidity : {mean_FA}')
print(f'citric_acid : {mean_CA}')
print(f'chlorides :{mean_C}')
print(f'free_sulfur_dioxide : {mean_FSD}')
print(f'total_sulfur_dioxide : {mean_TSD}')
print(f'density : {mean_D}')
print(f'pH : {mean_pH}')
print(f'quality : {mean_Q}')
```

```python
mode_FA = df.fixed_acidity.mode()
mode_CA = df.citric_acid.mode()
mode_C = df.chlorides.mode()
mode_FSD = df.free_sulfur_dioxide.mode()
mode_TSD = df.total_sulfur_dioxide.mode()
mode_D = df.density.mode()
mode_pH = df.pH.mode()
mode_Q = df.quality.mode()
```

```python
print(mode_FA)
print(mode_CA)
print(mode_C)
print(mode_FSD)
print(mode_TSD)
print(mode_D)
print(mode_pH)
print(mode_Q)
```

```python
median_FA = df.fixed_acidity.median()
median_CA = df.citric_acid.median()
median_C = df.chlorides.median()
median_FSD = df.free_sulfur_dioxide.median()
median_TSD = df.total_sulfur_dioxide.median()
median_D = df.density.median()
median_pH = df.pH.median()
median_Q = df.quality.median()
```

```python
print(f'Fixed_Acidity : {median_FA}')
print(f'citric_acid : {median_CA}')
print(f'chlorides :{median_C}')
print(f'free_sulfur_dioxide : {median_FSD}')
print(f'total_sulfur_dioxide : {median_TSD}')
print(f'density : {median_D}')
print(f'pH : {median_pH}')
print(f'quality : {median_Q}')
```

#### To replace the null values, I went with the median, i felt the values of the median seemed more reasonable.

```python
df['fixed_acidity'] = df['fixed_acidity'].fillna(median_FA)
df['citric_acid'] = df['citric_acid'].fillna(median_CA)
df['chlorides'] = df['chlorides'].fillna(median_FA)
df['free_sulfur_dioxide'] = df['free_sulfur_dioxide'].fillna(median_FSD)
df['total_sulfur_dioxide'] = df['total_sulfur_dioxide'].fillna(median_TSD)
df['density'] = df['density'].fillna(median_D)
df['pH'] = df['pH'].fillna(median_pH)
df['quality'] = df['quality'].fillna(median_Q)
```
